```python
class Solution:
    def sortArrayByParity(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        if len(A) ==0 or len(A) > 5000:
            print("Array exceeds max size")
            return None
        
        ArrayB = []
        for val in A:
            if val <0 or val > 5000:
                print("Bad value in array. Abort the function")
                return None
            
            if val % 2 == 0:
                ArrayB.insert(0, val)
            else:
                ArrayB.append(val)
        return ArrayB

sol = Solution()
array=[1,4,500000,3,2]
print (sol.sortArrayByParity(array))
```

```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: void Do not return anything, modify nums1 in-place instead.
        """
        idx1 = 0
        idx2 = 0
        temparray = []
        
        # go through the m elements in nums1 and n elements in nums2 to merge
        while True:
            if nums1[idx1] <= nums2[idx2]:
                temparray.append(nums1[idx1])
                idx1 += 1
            else:
                temparray.append(nums2[idx2])
                idx2 += 1

            #if exceed m elements in nums1 or n elements in nums2, exit the loop
            if idx1 >= m or idx2 >= n:
                break
            else:
                continue

        if idx1 < m:
            for ii in range(idx1,m):
                temparray.append(nums1[ii])
        else:
            for ii in range(idx2, n):
                temparray.append(nums2[ii])
        nums1.clear()
        for elem in temparray:
            nums1.append(elem)

sol = Solution()
nums1=[1,2,3,0,0,0]
nums2=[2,5,6]
sol.merge(nums1, 3, nums2, 3)
print(nums1)
```

```python
class Solution:
    def canPlaceFlowers(self, flowerbed, n):
        """
        :type flowerbed: List[int]
        :type n: int
        :rtype: bool
        """
        size = len(flowerbed)
        idx = 0
        totalspots = 0
        while True:
            # start counting plantable spots until hitting the end of array
            if flowerbed[idx] == 0:
                if (size == 1) or (idx == 0 and flowerbed[idx+1] == 0) or (idx == size-1 and flowerbed[idx-1]==0) or (flowerbed[idx-1] == 0 and flowerbed[idx+1] == 0):
                    totalspots += 1            
            if idx >= size - 1:
                break
            
            idx +=1
            continue

        if n <= totalspots:
            return True
        else:
            return False
        
flowerbed=[0,0,1]
sol = Solution()
print(sol.canPlaceFlowers(flowerbed, 1))
```

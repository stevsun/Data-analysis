```python
class MyHashMap:
    def __init__(self, size=100000):
        """
        Initialize your data structure here.
        """
        self.size = size
        self._keys = [None]*self.size
        self._values = [None]*self.size
        
    def hashFunc(self, key):
        return key % self.size
    
    def rehashFunc(self, key):
        return (key+1) % self.size

    def find(self, key):
        """
        Returns the position in table where key is, or -1 if not found
        :type key: int
        :rtype: int
        """
        startpos = self.hashFunc(key)
        found = False
        stop = False
        idx = startpos
        
        while self._keys[idx] != None and not found and not stop:
            if self._keys[idx] == key:
                found = True
            else:
                idx = self.rehashFunc(idx)
                if idx == startpos:
                    idx = -1
                    stop = True

        if self._keys[idx] == None: idx = -1
            
        return idx
    
    def put(self, key, value):
        """
        value will always be non-negative.
        :type key: int
        :type value: int
        :rtype: void
        """
        idx = self.hashFunc(key)
        if self._keys[idx] == None:
            self._keys[idx] = key
            self._values[idx] = value
        else:
            if self._keys[idx] == key:
                self._values[idx] = value
            else:
                nextidx = self.rehashFunc(idx)
                while self._keys[nextidx] != None and \
                    self._keys[nextidx] != key:
                    nextidx = self.rehashFunc(nextidx)
                if self._keys[nextidx] == None:
                    self._keys[nextidx] = key
                    self._values[nextidx] = value
                else:
                    self._values[nextidx] = value

    def get(self, key):
        """
        Returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key
        :type key: int
        :rtype: int
        """
        pos = self.find(key)
        print("get(): pos=", pos)
        if pos >= 0:
            return self._values[pos]
        else:
            return -1
        
    def remove(self, key):
        """
        Removes the mapping of the specified value key if this map contains a mapping for the key
        :type key: int
        :rtype: void
        """
        pos = self.find(key)
        
        if pos >= 0:
            self._keys[pos] = None
            self._values[pos] = None

        return

# Your MyHashMap object will be instantiated and called as such:
# obj = MyHashMap()
# obj.put(key,value)
# param_2 = obj.get(key)
# obj.remove(key)
```

```python
import requests
from bs4 import BeautifulSoup
import time

# TODO: Implement the continue_crawl function described above
def continue_crawl(search_history, target_url):
    print(search_history[-1])
    if search_history[-1] == target_url:
        print("Found the target URL, end the crawl")
        return False
    elif len(search_history) > 25:
        print("Reached the max iteration")
        return False
    elif search_history[-1] in search_history[:-1]:
        print("Cycling to previously visited page")
        return False
    elif len(search_history) != len(set(search_history)):
        print("Visited the same page before")
        return False
    else:
        print("New page, continue the crawling")
        return True
    

# TODO: implement function to find the first url link in the page    
def findFirstLink(url):
    page = requests.get(url)
    html = page.text
    soup = BeautifulSoup(html, 'html.parser')
    first_link = ""
    
    content_div = soup.find(id="mw-content-text").find(class_="mw-parser-output")
    for element in content_div.find_all("p", recursive=False):
        if element.find("a", recursive=False):
            first_relative_link = element.find("a", recursive=False).get('href')
            break
        
        
    tags = soup.find(id='mw-content-text').find(class_="mw-parser-output")
    
    for elements in tags.find_all("p", recursive = False):
        if elements.find("a", recursive=False) != None:
            first_link = elements.find("a", recursive=False).get('href')
            if first_link != None:
                return "https://en.wikipedia.org" + first_link
    return None

search_history = ['https://en.wikipedia.org/wiki/Special:Random']
while continue_crawl(search_history, 'https://en.wikipedia.org/wiki/Philosophy') == True:
    first_link = findFirstLink(search_history[-1])
    if first_link != None:
        search_history.append(first_link)
        time.sleep(3)
        continue
    else:
        break

print(search_history)
```
